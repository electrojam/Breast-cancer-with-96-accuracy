```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creamos array con datos aleatorios con distribución normal
# Datos de alturas de personas
# con una media = 170, std = 10 , 200 datos.

data = np.random.normal(170, 10, 200)
print(data)
```

```python
# Graficamos histrogram de data

# Creamos el gráfico histograma
plt.hist(data, color='salmon', edgecolor='black', alpha=0.6)

# Etiquetamos
plt.title('Distribución de alturas de personas')
plt.xlabel('Altura (cm)')
plt.ylabel('Densidad')

# Mostramos el gráfico
plt.show()
```

```python
# Modificamos divisiones, número de intervalos del gráfico histograma

# Creamos el gráfico histograma
plt.hist(data, color='salmon', edgecolor='black', alpha=0.6, bins=20)

# Etiquetamos
plt.title('Distribución de alturas de personas')
plt.xlabel('Altura (cm)')
plt.ylabel('Densidad')

# Mostramos el gráfico
plt.show()
```

```python
# Gráfico de Boxplot

# Extablecer una semilla de inicio
# para que no cambie la data cada que se ejecute
np.random.seed(0)

# Crear lista con 3 arrays con información de rango de edades
ages = [np.random.normal(30, 5, 100),
        np.random.normal(40, 5, 100),
        np.random.normal(35, 5, 100)]

print(ages)
```

```python
# Creamos gráfico Boxplot con los 3 datas
plt.boxplot(ages, patch_artist=True, notch=True, vert=True, labels=['Grupo', 'Grupo 2', 'Grupo 3'])

# Etiquetamos el gráfico
plt.title('Distribución de edades por grupo')
plt.xlabel('Grupo')
plt.ylabel('Edad (años)')

# Mostramos el gráfico
plt.show()
```
